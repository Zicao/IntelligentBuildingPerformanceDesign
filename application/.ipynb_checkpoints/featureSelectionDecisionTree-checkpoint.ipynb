```python
from aibpd.data.preprocessing import PreprocessingCBECS
from aibpd.data.database import Database
import numpy as np
import pandas as pd
from sklearn import tree

#load the database
database=Database()
CBECS_DF=database.select()
```

```python
CBECS_DF.columns
```

```python
X=CBECS_DF.values[:,2:]
Y=CBECS_DF.values[:,1]
m=CBECS_DF.shape[0]
n=int(m*0.8)
```

```python
clf_dt=tree.DecisionTreeClassifier(criterion='entropy')
clf_dt.fit(X[:n],Y[:n])
Y_dt=clf_dt.predict(X[n:m,:])
print(clf_dt.score(X[n:m,:],Y[n:m]))
dot_data=tree.export_graphviz(clf_dt,
                              feature_names=list(CBECS_DF.columns[2:]),
                              out_file=None,
                              filled=True,
                              rounded=True) 
```

```python
import graphviz
graph2 = graphviz.Source(dot_data)
```
